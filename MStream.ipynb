```python
#from plotting import *
#%matplotlib inline

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loading in data from Besla+2013 Model 1 

all = np.loadtxt('SMC_StarsPos_Model1.txt')
# x y z 
x = all[:,0] 
y = all[:,1]
z = all[:,2]  
r = np.sqrt(x**2 + y**2 +z**2)

# selecting stream stars
index = np.where(y > 0)
x1 = x[index]
y1 = y[index]
z1 = z[index]
r1 = r[index]

allv = np.loadtxt('SMC_StarsVel_Model1.txt')
# vx vy vz 
vx = allv[:,0] 
vy = allv[:,1]
vz = allv[:,2]  

v =np.sqrt(vx**2 + vy**2 + vz**2) 
v1 = v[index]
```

```python
# LMC  COM V and P

Lx = -1.06771
Ly = -41.04967
Lz = -27.83253
Lvx = -57.40
Lvy = -225.59
Lvz = 220.67
      
```

```python
# Radial Velocties

vrad = (vx*x + vy*y + vz*z )/r
vtan = np.sqrt(v**2 - vrad**2 )

vrad1 = vrad[index]
vtan1 = vtan[index]
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.subplot(111)

plt.scatter(x1, y1,c=vrad1) # , norm=matplotlib.colors.LogNorm())


#colorbar
cbar = plt.colorbar()
cbar.set_label('Vrad Galactocentric (km/s)', size=22)


#labels
plt.ylabel('Galactocentric y (kpc)', fontsize=22)
plt.xlabel('Galactocentric x (kpc)', fontsize=22)

plt.figtext(0.55, 0.83, 'Model 1', fontsize=22)

#adjust tick label font size
label_size = 15
matplotlib.rcParams['xtick.labelsize'] = label_size 
matplotlib.rcParams['ytick.labelsize'] = label_size


#set axis limits
plt.xlim(-70,50)
#plt.ylim(0,10)



#adjust tick label font size
label_size = 22
matplotlib.rcParams['xtick.labelsize'] = label_size 
matplotlib.rcParams['ytick.labelsize'] = label_size


ax.set_rasterized(True)
plt.savefig('MS_vrad.eps', rasterized=True, dpi=350)
```
